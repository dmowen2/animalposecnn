```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.utils import to_categorical
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import multilabel_confusion_matrix
import matplotlib.pyplot as plt
from statistics import mean

X = pickle.load(open("UNX.pickle", "rb"))
Y = pickle.load(open("UNY.pickle", "rb"))

X = X/255.0


Y = to_categorical(Y, 3)


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)



plt.title('Training and Validation accuracy')

history = None
traintimes = 30
total_acc_train = []
total_acc_test = []
while traintimes > 0:
    print("Train Iteration: " + str(traintimes))
    model=Sequential()
    model.add(Conv2D(255, (3,3), input_shape = X.shape[1:]))
    model.add(Activation("relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(64, (3,3)))
    model.add(Activation("relu"))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(128, (3,3)))
    model.add(Activation("relu"))
    model.add(BatchNormalization()) # adding the batch normalization here significantly improved the accuracy of the model
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(64, (3,3)))
    model.add(Activation("relu"))
    model.add(BatchNormalization()) # adding the batch normalization here significantly improved the accuracy of the model
    model.add(MaxPooling2D(pool_size=(2,2)))
    

    model.add(Conv2D(128, (3,3)))
    model.add(Activation("relu"))
    model.add(BatchNormalization()) # adding the batch normalization here significantly improved the accuracy of the model
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(64))
    model.add(Dense(32))
    model.add(Dense(3))
    model.add(Activation('softmax'))

    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

    # set back to 100
    history = model.fit(X_train, y_train, batch_size=32, epochs=100, validation_split=.2)


    
    _, train_acc = model.evaluate(X_train, y_train, verbose=0)
    _, test_acc = model.evaluate(X_test, y_test, verbose=0)
    
    y_pred=model.predict(X_test) 
    y_pred=np.argmax(y_pred, axis=1)
    y_nexttest=np.argmax(y_test, axis=1)
    cm = confusion_matrix(y_nexttest, y_pred)
    print(cm)
    
    
    
    # Plotting accuracies
    print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
    acc_train = history.history['accuracy']
    acc_val = history.history['val_accuracy']
    total_acc_train.append(train_acc)
    total_acc_test.append(test_acc)
    # switch back to 100
    epochs = range(0, 100)
    plt.plot(epochs, acc_train, 'g', label='Training accuracy')
    plt.plot(epochs, acc_val, 'b', label='validation accuracy')
    plt.show()
    # print(confusion_matrix(X_test, preds, labels=[1,2,3]))
    
    
    
    traintimes -= 1
plt.xlabel('Epochs')
plt.ylabel('Accuracy')


handle, label = plt.gca().get_legend_handles_labels()
labels = dict(zip(label, handle))
plt.legend(labels.values(), labels.keys())
plt.show()



plt.title('Training and Test accuracy per trial')
trials = range(0,31)
plt.plot(trials, total_acc_train, 'r', label="Training Accuracy")
plt.plot(trials, total_acc_test, 'b', label="Test Accuracy")

plt.xlabel('Trials')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


print("Average Train Acc: " + str(mean(total_acc_train)))
print("Average Test Acc: " + str(mean(total_acc_test)))




```
